## Library Import

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
from keras.layers import Conv2D, Dense, Dropout, Flatten, MaxPooling2D
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
```

## Inisiasi Data

```python
data_dir = 'Dataset/training'
target_size = (128,128)
classes = 26
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.1,
                                   horizontal_flip=False,
                                   validation_split=0.2)

train_gen = train_datagen.flow_from_directory(data_dir, target_size=target_size,
                                             shuffle=True,
                                             batch_size= batch_size,
                                             color_mode='grayscale',
                                             class_mode='categorical',
                                             subset='training')

val_gen = train_datagen.flow_from_directory(data_dir, target_size=target_size,
                                             batch_size= batch_size,
                                             color_mode='grayscale',
                                             class_mode='categorical',
                                             subset='validation')
```

```python
epoch = 50
train_size = len(train_gen)
steps_per_epoch = 32
validation_steps = 9

print(steps_per_epoch)
```

```python
x,y = train_gen.next()
for i in range(0,3):
    image = x[i]
    label = y[i]
    print (label)
    plt.imshow(image)
    plt.show()
```

## Membuat Model

Inisialisasi CNN

```python
SibiKlasifikasi = Sequential()
```

Convolution

```python
SibiKlasifikasi.add(Conv2D(32, (3, 3), input_shape=(128,128,1), activation='relu'))
SibiKlasifikasi.add(Conv2D(32, (3, 3), activation='relu'))
SibiKlasifikasi.add(MaxPooling2D(pool_size=(2, 2)))  # 2x2 is optimal
```

```python
SibiKlasifikasi.add(Conv2D(128, (3, 3), activation='relu'))
SibiKlasifikasi.add(MaxPooling2D(pool_size=(2, 2)))
SibiKlasifikasi.add(Dropout(0.5))
```

```python
SibiKlasifikasi.add(Conv2D(256, (3, 3), activation='relu'))
SibiKlasifikasi.add(MaxPooling2D(pool_size=(2, 2)))
SibiKlasifikasi.add(Dropout(0.5))
```

Flattening

```python
SibiKlasifikasi.add(Flatten())
```

Full Connection

```python
#SibiKlasifikasi.add(Dense(units=1000, activation='relu'))
SibiKlasifikasi.add(Dropout(0.5))
SibiKlasifikasi.add(Dense(units = 128, activation='relu'))
SibiKlasifikasi.add(Dense(units = 26, activation = 'softmax'))
```

```python
SibiKlasifikasi.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

SibiKlasifikasi.summary()
```

## Train Model

```python
##history = SibiKlasifikasi.fit(
    ##steps_per_epoch= 26,
    ##epochs=10,
    ##validation_data=val_gen,
    ##validation_steps=26)
hsitory = SibiKlasifikasi.fit(train_gen,
                         steps_per_epoch = steps_per_epoch,
                         epochs = epoch,
                         validation_data = val_gen,
                         validation_steps = validation_steps)
```

```python
acc = hsitory.history['accuracy']
val_acc = hsitory.history['val_accuracy']
loss = hsitory.history['loss']
val_loss = hsitory.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
